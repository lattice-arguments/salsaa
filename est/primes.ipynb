```python
N0 = 128
q = 2^50
for i in range(10000):
    q = q-1
    if not q.is_prime():
        continue
    if (q - 1) % N0 != 0:
        continue
    if (q - 1) % (2 * N0) == 0:
        continue
    q0 = q
    print(q)
    break
    # if q - 1)
```
